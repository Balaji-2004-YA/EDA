```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.histplot(data["Age_group"], kde=True)
plt.title("Distribution of Age_group")
plt.show()
```

```python
sns.histplot(data["Gender"], kde=True)
plt.title("Distribution of Gender ")
plt.show()
```

```python
data.head()
```

```python
sns.scatterplot(x="Age", y="Gender", data=data)
plt.title(x="Age", y="Gender")
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.boxplot(x="Age", y="Gender", data=data)
plt.xticks(rotation=90)
plt.title("age and gender")
plt.show()
```
